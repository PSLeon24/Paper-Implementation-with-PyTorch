```python
import matplotlib.pyplot as plt
import numpy as np

original_image = np.random.rand(100, 100)

plt.imshow(original_image, cmap='gray')
plt.title("original image(100x100)")
plt.axis('off')
plt.savefig("original_image.png")
plt.show()
```

```python
# 패치 크기
patch_size = 10

# 격자 방식 패치 그리기
fig, ax = plt.subplots()
ax.imshow(original_image, cmap='gray')

# 패치 그리드 그리기
for i in range(0, 100, patch_size):
    for j in range(0, 100, patch_size):
        rect = plt.Rectangle((j, i), patch_size, patch_size, edgecolor='red', facecolor='none')
        ax.add_patch(rect)

plt.title("grid patches (10x10)")
plt.axis('off')
plt.savefig("grid_patches.png")
plt.show()
```

```python
# 슬라이딩 윈도우 패치 그리기
fig, ax = plt.subplots()
ax.imshow(original_image, cmap='gray')

# 슬라이딩 윈도우 패치 그리기 (5 픽셀씩 겹침)
stride = 5
for i in range(0, 100 - patch_size + 1, stride):
    for j in range(0, 100 - patch_size + 1, stride):
        rect = plt.Rectangle((j, i), patch_size, patch_size, edgecolor='blue', facecolor='none')
        ax.add_patch(rect)

plt.title("sliding_window_patches (10x10)")
plt.axis('off')
plt.savefig("sliding_window_patches.png")
plt.show()
```
